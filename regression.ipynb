# Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.plot(x='LotArea', y='SalePrice', style='o')
```

```python
def net_input(W, X, b):
    return np.dot(W, X) + b
```

```python
W = 1
b = 0
dumb_X = np.arange(0, 150000)
dumb_Y = net_input(W, dumb_X, b)
```

```python
df.plot(x='LotArea', y='SalePrice', style='o')
plt.plot(dumb_X, dumb_Y, 'r--')
plt.show()
```

```python
def cost_function(W, X, b, Y):
    return np.sum(np.square(Y - net_input(W, X, b))) / X.shape[0]
```

```python
W = 1.1
b = 0

plt.plot(dumb_X, dumb_Y, 'bo')
plt.plot(dumb_X, net_input(W, dumb_X, b), 'r--')
plt.show()
cost_function(W, dumb_X, b, dumb_Y)
```

```python
def cost_function_derivative_w(W, X, b, Y):
    return 2 * np.sum((net_input(W, X, b) - Y) * X) / X.shape[0]
```

```python
def cost_function_derivative_b(W, X, b, Y):
    return 2 * np.sum(net_input(W, X, b) - Y) / X.shape[0]
```

```python
W = np.float64(0)
b = np.float64(0)

X = df['LotArea']
Y = df['SalePrice']

LEARNING_RATE = 10e-10
NUM_ITERATIONS = 50

for i in range(0, NUM_ITERATIONS):
    #print('{} step'.format(i))
    Y_hat = net_input(W, X, b)
    #print('cost function: ', cost_function(W, X, b, Y))
    #print('cost_function_derivative_w', cost_function_derivative_w(W, X, b, Y))
    W = W - LEARNING_RATE * cost_function_derivative_w(W, X, b, Y)
    b = b - LEARNING_RATE * cost_function_derivative_b(W, X, b, Y)
    
print(W)
print(b)
```

```python
df.plot(x='LotArea', y='SalePrice', style='o')
plt.plot(X, net_input(W, X, b), 'r--')
plt.ylim(0, 1000000)
plt.show()
```
